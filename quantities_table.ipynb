```python
import pandas as pd
```

```python
# Fill these values
file_path = '180 MAQ Apportionment Report.xlsx'
control_area = "3rd Floor East"
room = "0333"
```

```python
# Load the Excel file into a pandas dataframe
data = pd.read_excel(file_path)

# Filter out rows where "Room - Sublocation" equals "Total"
filtered_data = data[data["Room - Sublocation"] != "Total"]

# Display the first few rows of the filtered data
print(filtered_data.head())
```

```python
# Update the "CAS #" column to keep only the first value when there are multiple values
filtered_data.loc[:, "CAS #"]  = filtered_data["CAS #"].str.split(";").str[0]

# Display the first few rows to confirm the changes
print(filtered_data.head())
```

```python
filtered_data.loc[:, "Room - Sublocation"]  = filtered_data["Room - Sublocation"].str.split(" -").str[0]

# Display the first few rows to confirm the changes
print(filtered_data.head())
```

```python
#Function that generates a data frame for a particular control area

def filter_by_control_area(filtered_data, control_area):
    """
    Filters the dataframe for rows with the specified room number.

    Args:
        filtered_data (pd.DataFrame): The cleaned dataframe
        room_number (str): The control area to filter by

    Returns:
        pd.DataFrame: A dataframe of just the specificed control area
    """
    # Filter the dataframe for the specified room number
    control_area_df = filtered_data[filtered_data["Control Area"] == control_area]
    
    # Return the dataframe
    return control_area_df
```

```python
#Usage
contral_area_output = filter_by_control_area(filtered_data, control_area)
contral_area_output.head()
```

```python
#Function that generates a data frame per a particular room
#This is new version that saves different columns for the other tables

def filter_by_room(input_df, room_location):
    """
    Filters the dataframe for rows with the specified room number.

    Args:
        filtered_data (pd.DataFrame): The dataframe for the particular control area
        room_number (str): The room number to filter by.

    Returns:
        pd.DataFrame: A dataframe with the relevant columns for the specified room number.
    """
    # Filter the dataframe for the specified room number
    room_df = input_df[input_df["Room - Sublocation"] == room_location]
    
    # Return the dataframe for this room
    return room_df
```

```python
#Usage
control_area_data = contral_area_output
room_output_df = filter_by_room(control_area_data, room)
room_output_df.head()
```

```python
# Now take the output from that particular room and make the table for each hazard

# First, get all the hazards in this room
unique_hazard_codes = room_output_df["Fire Code Hazard"].unique()
print(unique_hazard_codes)

# Dictionary to store dataframes for each hazard code with specific columns
hazard_code_dataframes = {}

# Columns to keep in the new dataframes
columns_to_keep = ["Chemical Name", "CAS #", "US Units", "In Storage"]

# Loop over each unique hazard code
for hazard_code in unique_hazard_codes:
    # Filter rows with the current hazard code
    filtered_rows = room_output_df[room_output_df["Fire Code Hazard"] == hazard_code]
    
    # Select only the specified columns
    filtered_rows = filtered_rows[columns_to_keep]
    
    # Check if all "US Units" values are the same
    unique_units = filtered_rows["US Units"].unique()
    
    if len(unique_units) == 1:
        # All "US Units" are the same, store as one dataframe
        hazard_code_dataframes[hazard_code] = filtered_rows
    else:
        # Split into separate dataframes based on "US Units"
        for unit in unique_units:
            sub_filtered_rows = filtered_rows[filtered_rows["US Units"] == unit]
            hazard_code_dataframes[f"{hazard_code} - {unit}"] = sub_filtered_rows


# Example: Accessing a specific dataframe by hazard code
for hazard, df in hazard_code_dataframes.items():
    print(f"Hazard Code: {hazard} (Entries: {len(df)})")
    print(display(df))  # Show the first few rows for each dataframe

```

```python
# Now this is the part that writes everything to Excel
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows

# Create a new Excel workbook
output_file = "quantities_report.xlsx"
wb = Workbook()

# Remove the default sheet created in the workbook
default_sheet = wb.active
wb.remove(default_sheet)

# Control area and room information
# The control_area variable is defined above
control_area_for_report = control_area
# The room variable is defined above
room_number_for_report = room

# Add a sheet for the report
sheet = wb.create_sheet(title="Hazard Report")

# Write the control area and room number at the top
sheet.append([f"Control Area: {control_area_for_report}"])
sheet.append([f"Room: {room_number_for_report}"])
sheet.append([])  # Add a blank row

# Loop through hazard codes and write each section
for hazard_code, df in hazard_code_dataframes.items():
    # Write the hazard code as a section header
    sheet.append([f"Hazard Code: {hazard_code}"])
    
    # Write the dataframe below the hazard code
    for row in dataframe_to_rows(df, index=False, header=True):
        sheet.append(row)

    # Calculate and print the total for the "In Storage" column
    total_in_storage = df["In Storage"].sum()
    sheet.append([f"Summary for Table: {total_in_storage:.3f}"])  # Write total to the Excel sheet
    
    # Add a blank row after each section for readability
    sheet.append([])

# Save the workbook to a file
wb.save(output_file)

print(f"Hazard report saved to {output_file}")
```
